# Shopping Cart Program

Create a shopping cart with the following basic functionality:
1. Display the list of items available in the store, along with the corresponding quantities currently in stock and the cost per item.
2. Ask the customer to enter the number of quantities required for each item.
3. Display a message if the item is out of stock.
4. Accept customer name and delievery address as the input.
5. Calculate delievery charges based on the distance of the delievery address from the store, as shown below in $.
* 10 for up to 5 miles.
* 12 between 5-10 miles.
* 14 between 10-15 miles.
* 20 for beyond 15 miles.
6. Display the final Bill with the items purchased and the customer details.
7. The program should keep running in the infinite loop and prompt the user to enter "yes" to continue and enter "no" to terminate the program.
8. The items sold to a customer must automatically be updated to display the correct stock levels to the next customer.

```python
print("%51s"%"GROSSERY STORE\n")

def Essential():
    print("%55s"%("ESSENTIAL ITEMS AVAILABLE"))
    print("%15s %18s %27s %16s"%("Sr. No.", "Items", "Available_Quantity", "Price (in $)"))
    print("%12d %30s %10d %19.2f"%(1, 'Biscuits: a  pack of 20', 5, 2.50))
    print("%12d %30s %10d %19.2f"%(2, 'Cereals: 2 pounds/pack', 10, 3.50))
    print("%12d %30s %10d %19.2f"%(3, 'Chicken: 1 pounds/pack', 20, 13.50))
    print("%12d %30s %10d %19.2f"%(4, 'Oats: 4 pounds/pack', 20, 3.50))
    print("%12d %30s %10d %19.2f"%(5, 'Rice: 2 pounds/pack', 30, 4.50))
    print('\n')        

    Biscuits = int(input("How much packets of Biscuits are required: "))
    Cereals = int(input("How much packets of Cereals are required: "))
    Chicken = int(input("How much packets of frozen Chicken are required: "))
    Oats = int(input("How much packets of Oats are required: "))
    Rice = int(input("How much packets of Rice are required: "))
    print('\n')
    return Biscuits, Cereals, Chicken, Oats, Rice

def Details():
    print("Fill in your Details:")
    Name = input("Enter your Name: ")
    Address = input("Enter your Address: ")
    Distance = int(input("Distance is: "))
    print("Distance from Shopping Mart (in miles) is: {}".format(Distance),"miles")
    print('\n')
    return Name, Address, Distance

def Bill(Distance):
    delievery_charges = 0
    if Distance < 5:
        delievery_charges = 10
    elif Distance > 5 and Distance < 10:
        delievery_charges = 12
    elif Distance > 10 and Distance < 15:
        delievery_charges = 14
    elif Distance > 15:
        delievery_charges = 20
    else:
        delievery_charges= 0
    return delievery_charges

def Total_price():
    Total_Price =  Biscuits*2.50 + Cereals*3.50 + Chicken*13.50 + Oats*3.50 + Rice*4.50
    Total_Amount = Total_Price + delievery_charges
    return Total_Price, Total_Amount

def Final_Bill():
    print("%41s"%"BILL")
    print("%15s %13s %17s %17s %15s"%("SR. NO", "Items", "Quantity", "Product_Price", "TotalCost"))
    print("%60s %19s"%("(in $)", "(in $)"))
    print("%12d %17s %12d %15.2f %17.2f"%(1, "Biscuits", Biscuits, 2.50, Biscuits*2.50))
    print("%12d %17s %12d %15.2f %17.2f"%(2, "Cereals", Cereals, 3.50, Cereals*3.50))
    print("%12d %17s %12d %15.2f %17.2f"%(3, "Chicken", Chicken, 13.50, Chicken*13.50))
    print("%12d %17s %12d %15.2f %17.2f"%(4, "Oats", Oats, 3.50, Oats*3.50))
    print("%12d %17s %12d %15.2f %17.2f"%(5, "Rice", Rice, 4.50, Rice*4.50))
    print('\n')

while True:
    print("Do you want to continue shopping?:")
    print("Yes/No")
    Choice = input("Enter your Choice Yes or No: ")
    print('\n')
    
    if Choice == 'Yes' or Choice == 'yes' or Choice == 'YES':
        Biscuits, Cereals, Chicken, Oats, Rice = Essential()
        Name, Address, Distance = Details()
        delievery_charges = Bill(Distance)
        Total_Price, Total_Amount = Total_price()
        Final_Bill()
        print("\nTotal Price with Items is: $",Total_Price)
        print("Total Amount with delievery charges is: $",Total_Amount)
        print('\n')
        
        print("Do you want to Continue Shopping?:")
        print("Yes/No")
        Choice = input("Enter your Choice Yes or No: ")
        print('\n')
        print("--"*30)
        print("THANK YOU")
        print("Have a Nice Day")
        print("--"*30,'\n')
        continue
    elif Choice == 'No' or Choice == 'no' or Choice == 'NO':
        break
    else:
        print("Wrong Choice Choosen:", Choice)
        print("Please Choose Correct Choice or Your Shopping is Done\n")
        continue
    break    
```
