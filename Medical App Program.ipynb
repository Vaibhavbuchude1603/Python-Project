```python
print("%51s"%"MEDICAL APP PROGRAM\n")
print("Hello")
print("Welcome to our Medical App")
print("--"*30,'\n')

def Products():
    print("%37s"%("MEDICINES IN MEDICAL"))
    print("%15s %15s %15s"%("Sr. No.", "Medicines", "MRP (in Rs)"))
    print("%11d %18s %13.2f"%(1, 'Crocine', 2.50))
    print("%11d %18s %13.2f"%(2, 'DOLO 650', 5.40))
    print("%11d %18s %13.2f"%(3, 'Thyronorm', 399.33))
    print("%11d %18s %13.2f"%(4, 'Combiphlem', 2.00))
    print("%11d %18s %13.2f"%(5, 'Zandu Balm', 45.25))
    print('\n')
          
    Crocine = int(input("How many quantity of Crocine you required: "))
    Dolo_650 = int(input("How many quantity of Dolo 650 you required: "))
    Thyronorm = int(input("How many quantity of Thyronorm you required: "))
    Combiphlem = int(input("How many quantity of Combiphlem you required: "))
    Zandu_Balm = int(input("How many quantity of Zandu Balm you required: "))
    print('\n')
    return Crocine, Dolo_650, Thyronorm, Combiphlem, Zandu_Balm
          
def Details():
    print("Fill your Details/Information:")
    Name = input("Enter your Full Name: ")
    Address = input("Enter your Address: ")
    Distance = int(input("Distance is: "))
    print("Distance from Medical (in KM) is: {}".format(Distance),"KM")
    print('\n')
    return Name, Address, Distance

def Charges(Distance):
    delievery_charges = 0
    if Distance <= 2:
        delievery_charges = 50
    elif Distance > 2 and Distance <= 10:
        delievery_charges = 100
    elif Distance > 10 and Distance <= 20:
        delievery_charges = 200
    elif Distance > 20 and Distance <= 30:
        delievery_charges = 300
    elif Distance > 30 and Distance <= 40:
        delievery_charges = 400
    elif Distance > 40:
        delievery_charges = 550
    else:
        delievery_charges = 0
    return delievery_charges
          
def Price():
    Total_Price = Crocine*2.50 + Dolo_650*5.40 + Thyronorm*399.33 + Combiphlem*2.00 + Zandu_Balm*45.25
    Total_Amount = Total_Price + delievery_charges
    return Total_Price, Total_Amount
          
def Bill():
    print("%50s"%"YOUR BILL")
    print("%15s %16s %18s %11s %16s"%("Sr. No", "Medicines", "Quantity", "Price", "TotalCost"))
    print("%12d %18s %15d %15.2f %15.2f"%(1, "Crocine", Crocine, 2.50, Crocine*2.50))
    print("%12d %18s %15d %15.2f %15.2f"%(2, "Dolo 650", Dolo_650, 5.40, Dolo_650*5.40))
    print("%12d %18s %15d %15.2f %15.2f"%(3, "Thyronorm", Thyronorm, 399.33, Thyronorm*399.33))
    print("%12d %18s %15d %15.2f %15.2f"%(4, "Combiphlem", Combiphlem, 2.00, Combiphlem*2.00))
    print("%12d %18s %15d %15.2f %15.2f"%(5, "Zandu Balm", Zandu_Balm, 45.25, Zandu_Balm*45.25))
    print('\n')
          
while True:
    print("Do you want to continue Shopping?:")
    print("Yes/No")
    Choice = input("Enter your Choice Yes or No: ")
    print('\n')
          
    if Choice == 'Yes' or Choice == 'yes' or Choice == 'YES':
        Crocine, Dolo_650, Thyronorm, Combiphlem, Zandu_Balm = Products()
        Name, Address, Distance = Details()
        delievery_charges = Charges(Distance)
        Total_Price, Total_Amount = Price()
        Bill()
        print("\nTotal Price with Medicines is: Rs",Total_Price)
        print("Total Amount with Delievery charges is: Rs",Total_Amount, '\n')
        
        print("Do you want to Continue Shopping?:")
        print("Yes/No")
        Choice = input("Enter your Choice Yes or No: ")
        print('\n')
        print("--"*30)
        print("THANK YOU")
        print("Have a Nice Day")
        print("--"*30, '\n')
        continue
    elif Choice == 'No' or Choice == 'no' or Choice == 'NO':
        break
    else:
        print("Wrong Choice Choosen:", Choice)
        print("Please Choose Correct Choice\n")
        continue
    break
```
